```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import os


# Columns 
# 1: id (int) [0-...]
# 2: year (number)
# 3: month (number)
# 4: day_of_month (number)
# 5: prp (float) [mm]
# 6: tmax (float) [°C]
# 7: tmin (float) [°C]
# negative values (-99.9) are missing data


# Carga de datos
df = pd.read_csv(os.path.join('..','..', 'data','data_precipitaciones', 'SANMARTIN_NARANJILLO.csv'))
# Adding column names
df.columns = ['id', 'year', 'month', 'day_of_month', 'prp', 'tmax', 'tmin']
# replacing negative values lower than -90 with NaN
df[df<-90] = np.nan
```

```python
# make an histogram of the precipitation data
df['prp'].hist(bins=30)
plt.show()
```

```python
# histogram of the temperature data
df['tmax'].hist(bins=30)
# leyend
plt.xlabel('Temperature (°C)')
plt.ylabel('Frequency')
plt.show()
```
